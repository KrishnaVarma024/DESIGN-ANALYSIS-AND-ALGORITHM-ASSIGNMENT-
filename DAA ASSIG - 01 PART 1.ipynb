```python
# PART - 02
#19CSE302  Design and Analysis of Algorithms          Lab Assignment 2

# Q1 : Find the sum of first N natural numbers using Iterative and Recursive algorithms. 
# Find the time taken to execute the same by varying Ns value and plot it using pythons plot function.

import time
import matplotlib.pyplot as plt
def iterative_sum(N):
    sum = 0
    for i in range(1, N+1):
        sum += i
    return sum
def recursive_sum(N):
    if N == 0:
        return 0
    else:
        return N + recursive_sum(N-1)
def measure_iterative_time(N):
    start_time = time.time()
    iterative_sum(N)
    end_time = time.time()
    return end_time - start_time
def measure_recursive_time(N):
    start_time = time.time()
    recursive_sum(N)
    end_time = time.time()
    return end_time - start_time
N = int(input("Enter the value of N: "))
iterative_sum_value = iterative_sum(N)
print(f"Sum of first {N} natural numbers (iterative): {iterative_sum_value}")
recursive_sum_value = recursive_sum(N)
print(f"Sum of first {N} natural numbers (recursive): {recursive_sum_value}")
iterative_times = [measure_iterative_time(i) for i in range(1, N+1)]
recursive_times = [measure_recursive_time(i) for i in range(1, N+1)]
plt.plot(range(1, N+1), iterative_times, label='Iterative')
plt.plot(range(1, N+1), recursive_times, label='Recursive')
plt.xlabel('N')
plt.ylabel('Time (seconds)')
plt.title('Time Taken to Compute Sum of First N Natural Numbers')
plt.legend()
plt.show()
```

```python
# Given m sorted lists/ arrays, each containing n elements, print them efficiently in sorted order

import random
import time
import matplotlib.pyplot as plt

def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1

def binary_search(arr, key):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == key:
            return mid
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1
    return -1

arr = [random.randint(1, 1000) for _ in range(10000)]

search_keys = [random.randint(1, 1000) for _ in range(5)]

linear_times = []
binary_times = []

for key in search_keys:
    start_time = time.time()
    linear_search(arr, key)
    end_time = time.time()
    linear_times.append(end_time - start_time)

    start_time = time.time()
    binary_search(arr, key)
    end_time = time.time()
    binary_times.append(end_time - start_time)

# Plotting
plt.plot(search_keys, linear_times, label='Linear Search')
plt.plot(search_keys, binary_times, label='Binary Search')
plt.xlabel('Search Key')
plt.ylabel('Time (seconds)')
plt.title('Comparison of Linear and Binary Search')
plt.legend()
plt.show()134
```

```python
# Given an array of size N, find the K largest elements in the array where K<<<N.

def string_to_integer(string):
    if len(string) == 0:
        return 0
    digit = int(string[0])
    remaining_string = string[1:]
    return digit + 10 * string_to_integer(remaining_string)

input_string = input("Enter a string of digits: ")
print("You entered:", input_string)

if input_string.isdigit():
    result = string_to_integer(input_string)
    print("Integer representation:", result)
else:
    print("Invalid input! Please enter only digits.")
```

```python
# Given a set of activities, along with the starting and finishing time of each activity, find the maximum number of activities 
# performed by a single person assuming that a person can only work on a single activity at a time.

def reverse_string(s):
    if len(s) == 0:
        return s
    else:
        return reverse_string(s[1:]) + s[0]

input_string = "pots&pans"
reversed_string = reverse_string(input_string)
print("Original string:", input_string)
print("Reversed string:", reversed_string)
```

```python
# Given a set of intervals, print all non-overlapping intervals after merging the overlapping intervals.

def is_palindrome(s):
    
    if len(s) <= 1:
        return True
    
    if s[0] == s[-1]:
        
        return is_palindrome(s[1:-1])
    else:
        return False

strings = ["racecar", "gohangasalamiimalasagnahog", "hello", "level"]
for s in strings:
    print(f"'{s}' is a palindrome: {is_palindrome(s)}")
```
